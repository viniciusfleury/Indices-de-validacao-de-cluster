#Minimo

```python
def δ(ck, ce_global):
    values = np.ones(len(ck))
    for i in range(0, len(ck)):
      ce = pd.DataFrame(ck[i])
      ce = ce.mean()
      ce = ce.to_numpy()
      values[i] = np.sqrt(np.sum(np.power(ce-ce_global, 2)))
    inter = np.min(values)
    return inter
```

```python
δ(k_list, ce_global)
```

#Maximo

```python
def Δ(ci):
    values = np.zeros(len(ci))
    ce = pd.DataFrame(ci)
    ce = ce.mean()
    ce = ce.to_numpy()
    for i in range(0, len(ci)):
      values[i] = np.sqrt(np.sum(np.power(ce-ci[i], 2)))
    return np.max(values)
```

Dunn

```python
def dunn(k_list, ce_global):
    δs = np.ones(len(k_list))
    Δs = np.zeros(len(k_list))
    d = []
    d2 = []
    δs = δ(k_list, ce_global)
    d.append(δs)
    for k in range(0, len(k_list)):
      Δs[k] = Δ(k_list[k])
      d2.append(Δs[k])
    di = np.min(d)/np.max(d2)
    return di
```

```python
%%time
dunn(k_list, ce_global)
```
