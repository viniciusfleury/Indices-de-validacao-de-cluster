```python
target = []

for i in range(0, 5000000):
  if i >=0 and i<1000000:
    target.append(0)
  elif i>=1000000 and i<2000000:
    target.append(1)
  elif i>=2000000 and i<3000000:
    target.append(2)
  elif i>=3000000 and i<4000000:
    target.append(3)
  else:
    target.append(4)
```

```python
k_list = []
for i in range(n_clusters):
  clus = prediction.copy()
  clus = clus.loc[clus.pred == i]
  clus.drop(["pred"], axis=1, inplace=True)
  k_list.append(clus.values)
```

#Calculando o centroid global

```python
ce_global = df.mean()
```

```python
ce_global = ce_global.to_numpy()
```

```python
print(ce_global)
```

#Coesão:

```python
def C_S(ck, ce_global):
    values = np.ones(len(ck))
    for i in range(0, len(ck)):
      ce = pd.DataFrame(ck[i])
      ce = ce.mean()
      ce = ce.to_numpy()
      values[i] = np.sqrt(np.sum(np.power(ce-ce_global, 2)))
    c_s = values.mean()
    return c_s
```

```python
C_S(k_list, ce_global)
```

#Separação:


```python
def S_S(ck):
  values = np.zeros(len(df))
  for i in range(0, len(ck)):
    ce = pd.DataFrame(ck[i])
    ce = ce.mean()
    ce = ce.to_numpy()
    for k in range(0, len(ck[i])):
     values[k] = np.sqrt(np.sum(np.power(ce-ck[i][k], 2)))
  s_s=values.mean()
  return s_s
```

```python
S_S(k_list)
```

#BD - Silhouette:

```python
def BD_S(k_list, ce_global):
  inter=C_S(k_list, ce_global)
  intra=S_S(k_list)
  max_bd_s=max(inter, intra)
  bd_s=(inter-intra)/max_bd_s
  return bd_s
```
